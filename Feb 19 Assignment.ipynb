#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

* Flask is a micro web framework written in Python that allows developers to build web applications quickly and easily. It is called a "micro" framework because it does not require particular tools or libraries, and it has no database abstraction layer, form validation, or any other components that are common in other web frameworks like Django. This simplicity makes it easy to learn and use, and also allows for greater flexibility in how developers choose to structure their applications.
* Advantages of Flask Framework:
    * Lightweight and flexible
    * Easy to learn
    * Quick to set up
    * Extensible
    * Compatible with other Python libraries
    * Well-documented

#### Q3. What is App routing in Flask? Why do we use app routes?

* In Flask, an app route is a URL pattern that is associated with a specific view function. When a user visits a particular URL, Flask maps that URL to the corresponding view function and then executes the function, generating a response that is returned to the user's browser.

* App routing is a core component of Flask, and it is used to define the different URL paths that your Flask application should respond to. You use app routes to create a mapping between a URL and a specific function or view in your Flask application. These routes define the different pages or actions that a user can take within your application.
---
### Example

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to our page!'

@app.route('/login')
def login():
    return 'This is login page'

@app.route('/signup')
def signup():
    return 'This is signup page'

if __name__=="__main__":
    app.run()
```

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

* In Flask, the **url_for()** function is used for URL building. This function generates a URL for a specified endpoint and any associated arguments based on the routing information defined in your Flask application.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to my website!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/contact')
def contact():
    return 'This is the contact page.'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('about'))
        print(url_for('contact'))
```
