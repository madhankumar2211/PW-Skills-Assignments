#### Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its application.


* **Min-Max scaling** is a data preprocessing technique used to rescale numerical data within a fixed range, usually between 0 and 1. This technique transforms the values of each feature in the dataset such that the minimum value is mapped to 0, and the maximum value is mapped to 1. All other values are scaled proportionally between these two points.
~~~
The formula for Min-Max scaling is:
~~~
> **X_scaled = (X - X_min) / (X_max - X_min)**

* where X is the original value, X_min is the minimum value of the feature, X_max is the maximum value of the feature, and X_scaled is the rescaled value.


* Min-Max scaling is used in data preprocessing to normalize the features of a dataset, which can help improve the performance of machine learning algorithms that are sensitive to the scale of the input data. For example, gradient descent optimization algorithms can converge faster when the features are on the same scale, and distance-based algorithms such as K-nearest neighbors can be affected by the scale of the features.


**Example**

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

X = np.array([[10., 2., 1.],
              [5., 8., 2.],
              [2., 7., 8.]])

# Instantiate the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the data
X_scaled = scaler.fit_transform(X)

# Print the rescaled data
print(X_scaled)
```

#### Q2. What is the Unit Vector technique in feature scaling, and how does it differ from Min-Max scaling? Provide an example to illustrate its application.


* The Unit Vector technique is a feature scaling method that rescales the values of each feature in a dataset to have a unit norm (i.e., a magnitude of 1). This technique is also known as normalization or L2 normalization.

* The main idea behind Unit Vector scaling is to scale each data point such that its feature vector has the same magnitude, which can be useful when the magnitude of the features is important, but their absolute values are not. This technique can also help improve the performance of some machine learning algorithms, particularly those that use distance measures or dot products between vectors, such as K-nearest neighbors, Support Vector Machines (SVMs), and neural networks.
~~~
The formula for Unit Vector scaling is:
~~~
> **X_unit = X / ||X||**

* where X is the original feature vector, ||X|| is the Euclidean norm (i.e., the magnitude) of the vector, and X_unit is the rescaled feature vector.

* Unit Vector scaling differs from Min-Max scaling in that it doesn't force the values of the features to a fixed range, but instead, it normalizes them to have the same magnitude. In other words, Unit Vector scaling preserves the direction of the original vectors but rescales them to have a unit length.

**Example:**

```python
from sklearn.preprocessing import Normalizer
import numpy as np

#sample dataset
X = np.array([[70., 1.8],
              [65., 1.75],
              [80., 1.9],
              [62., 1.68],
              [75., 1.83]])

# Instantiate the Normalizer
scaler = Normalizer()

# Fit and transform the data
X_scaled = scaler.fit_transform(X)

# Print the rescaled data
print(X_scaled)
```

#### Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an example to illustrate its application.


* Principal Component Analysis (PCA) is a widely used technique in dimensionality reduction that is used to transform high-dimensional data into a lower-dimensional space while retaining as much of the original variability as possible. In other words, PCA can help to identify the most important patterns or features in a dataset and project the data onto a lower-dimensional space based on these patterns.

* PCA works by identifying the principal components of a dataset, which are the directions in which the data varies the most. These principal components are orthogonal to each other and are ranked by the amount of variance they explain in the data. The first principal component explains the most variance, followed by the second, and so on. By projecting the data onto these principal components, we can reduce the dimensionality of the data while retaining the maximum amount of variability.

* PCA is particularly useful in situations where the number of features in a dataset is very high, making it difficult to visualize or analyze the data effectively. By reducing the dimensionality of the data, we can often gain insights into the underlying structure of the data and identify important patterns or relationships that might not be apparent in the original high-dimensional space.

**Example:**
-  we have a dataset containing the measurements of 1000 flowers, each with 4 features: sepal length, sepal width, petal length, and petal width. We want to apply PCA to the data to reduce its dimensionality and identify the most important patterns or features. Here's how we can do it:

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Instantiate a PCA object
pca = PCA(n_components=2)

# Fit and transform the data
X_pca = pca.fit_transform(X)

# Plot the transformed data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=iris.target)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

#### Q4. What is the relationship between PCA and Feature Extraction, and how can PCA be used for Feature Extraction? Provide an example to illustrate this concept.


* PCA and feature extraction are closely related concepts in machine learning. Feature extraction is the process of transforming the raw input data into a new set of features that are more representative of the underlying patterns in the data. PCA is one method for performing feature extraction, where the new set of features is constructed as a linear combination of the original features.

* PCA can be used for feature extraction in the following way:
~~~
1. Collect a dataset of N samples, where each sample has M features.
2. Standardize the data to have zero mean and unit variance across all features.
3. Compute the covariance matrix of the standardized data.
4. Compute the eigenvectors and eigenvalues of the covariance matrix.
5. Choose the top K eigenvectors with the largest eigenvalues.
6. Project the data onto the K-dimensional subspace defined by the top K eigenvectors to obtain the new feature.
~~~

* The resulting new features are linear combinations of the original features, where each new feature is a weighted sum of the original features. The weights are given by the components of the corresponding eigenvectors, and the importance of each new feature is proportional to the corresponding eigenvalue.

**Example:**


```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the digits dataset
digits = load_digits()
X = digits.data

# Instantiate a PCA object
pca = PCA(n_components=2)

# Fit and transform the data
X_pca = pca.fit_transform(X)

# Plot the transformed data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=digits.target)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

    * You would then rank the principal components by the amount of variance they explain in the data. You can do this using the explained_variance_ratio_ attribute of the PCA object.

    * You would choose a number of principal components to keep based on the amount of variance they explain. A common approach is to keep enough principal components to explain 90% or 95% of the variance in the data.

    * Finally, you would transform the original dataset into the reduced-dimensionality dataset by selecting only the principal components you chose in step 4. This new dataset would have fewer columns than the original dataset and would be less computationally expensive to work with.

#### Q7. For a dataset containing the following values: [1, 5, 10, 15, 20], perform Min-Max scaling to transform the values to a range of -1 to 1.


```python
import numpy as np

# Define the original dataset
data = np.array([1, 5, 10, 15, 20])

# Perform Min-Max scaling to transform the data to a range of -1 to 1
scaled_data = (data - np.min(data)) / (np.max(data) - np.min(data)) * 2 - 1

# Print the scaled data
print(scaled_data)
```
